### Estatística Descritiva

```r
#Carregando a base de dados
dados <- read.csv("PelicanStores.csv", check.names=FALSE, fileEncoding="Windows-1252")
head(dados)
```

```r
#Resumo dos dados
print(summary(dados))
```

```r
# Carregando as bibliotecas
library(tidyverse)

# Renomeando as colunas
names(dados)[2] <- "TipoCliente"
names(dados)[4] <- "VendasLiquidas"
names(dados)[5] <- "MetodoPagamento"
names(dados)[6] <- "Genero"
names(dados)[7] <- "EstadoCivil"
print(head(dados))
```

```r
# Função para cálculo de moda - https://www.tutorialspoint.com/r/r_mean_median_mode.htm
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
# Convertendo os valores de Vendas Liquidas de texto para valores numéricos
dados$VendasLiquidas = gsub(",", ".", dados$VendasLiquidas)
dados$VendasLiquidas = as.numeric(as.character(dados$VendasLiquidas))
```

### Variáveis numéricas

```r
#Dados da variável Itens

print("Itens")
print(paste("Média: ", mean(dados$Itens)))
print(paste("Mediana: ", median(dados$Itens)))
print(paste("Desvio padrão:", sd(dados$Itens)))
print(paste("Variância:", var(dados$Itens)))
print(paste("Moda:", getmode(dados$Itens)))
```

```r
#Dados da variável VendasLiquidas

print("Vendas Liquidas")
print(paste("Média: ", mean(dados$VendasLiquidas)))
print(paste("Mediana: ", median(dados$VendasLiquidas)))
print(paste("Desvio padrão:", sd(dados$VendasLiquidas)))
print(paste("Variância:", var(dados$VendasLiquidas)))
print(paste("Moda:", getmode(dados$VendasLiquidas)))
```

```r
#Dados da variável Idade

print("Idade")
print(paste("Média: ", mean(dados$Idade)))
print(paste("Mediana: ", median(dados$Idade)))
print(paste("Desvio padrão:", sd(dados$Idade)))
print(paste("Variância:", var(dados$Idade)))
print(paste("Moda:", getmode(dados$Idade)))
```

### Variáveis categóricas
